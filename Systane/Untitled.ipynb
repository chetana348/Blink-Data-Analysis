```python
import pandas as pd
import plotly.graph_objects as go

# read in data from text file
data = pd.read_csv(r"C:\Users\cheta\Downloads\SP Codes\Systane\data\Akshaya_Right-17@09@41.TXT", delim_whitespace=True, header=None)

# create figure object
fig = go.Figure()

# add trace to figure
fig.add_trace(go.Scatter(x=data[0], y=data[1], mode='lines', name='Column 1 vs Column 2'))
fig.add_trace(go.Scatter(x=data[0], y=data[2], mode='lines', name='Column 1 vs Column 3'))
fig.add_trace(go.Scatter(x=data[0], y=data[3], mode='lines', name='Column 1 vs Column 4'))
fig.update_xaxes(rangeslider_visible=True)
# customize layout
fig.update_layout(title='Data Plot', xaxis_title='X-axis', yaxis_title='Y-axis')

# display figure
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# read in data from text file
data = pd.read_csv(r"C:\Users\cheta\Downloads\SP Codes\Systane\data\Nandhini_Right-15@56@46.TXT", delim_whitespace=True, header=None)

# create figure object
fig = go.Figure()

# add trace to figure for each column
for i in range(0, len(data.columns)):
    fig.add_trace(go.Scatter(x=data.index, y=data[i], mode='lines', name=f'Column {i}'))

# customize layout
fig.update_layout(title='Data Plot', xaxis_title='Index', yaxis_title='Value')
fig.update_xaxes(rangeslider_visible=True)

# display figure
fig.show()
```
